```python
from anastruct import SystemElements
ss = SystemElements(EA=5e3,EI=1e5)
```

```python
# insertar elementos
ss.add_element(location=[[0,0],[2,0]])
ss.add_element(location=[[2,0],[2.9,1.2]])

# insertar restricciones
ss.add_support_hinged(node_id=[1])
ss.add_support_rotational(node_id=3)

# cargas
ss.q_load(element_id=2,q=-7.5,direction='element')
```

```python
# mostrar estructura
ss.show_structure()
```

```python
# resolver
ss.solve()
```

```python
ss.show_reaction_force()
```
